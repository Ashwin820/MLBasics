```python
count = 0
for i in range(0,25):
    new_string = file[0][i].split()
    print(new_string)
    length = len(new_string)
    count += length
```

```python
print("No. of words in first column: ",count)
```

```python
d = {}
for i in string.punctuation:
    d[i] = " "
```

```python
unique_count = 0
unique_list = []
for i in range(0,25):
    string_1 = file[0][i]
    new_string_1 = string_1.translate(str.maketrans(d))
    string_without_punc = new_string_1.split()
    print(string_without_punc)
    for j in range(0,len(string_without_punc)):
        if string_without_punc[j] not in unique_list:
            unique_list.append(string_without_punc[j])
```

```python
print("No. of unique words in first column: ",len(unique_list))
```
